USING WEBCAM TO DETECT THE IMAGES IN REAL TIME

```python
import numpy
import cv2
from ultralytics import YOLO
```

```python
import os
print("Current directory:", os.getcwd())
```

```python
# train the weights "best"
model = YOLO('runs/detect/train2/weights/best.pt')
```

```python
results = model("images_nottrained/mouse-7.jpg")
results.show()
```

```python
# get annoted image because of the shape
annoted_img = results[0].plot()
```

```python
# show it using opencv
cv2.imshow("Detected_mouse", annoted_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite("detected_mouse-7.jpg", annoted_img)
```

```python
# using webcam
capt_entity = cv2.VideoCapture(0)   #initializes the webcam (1 or 2 will initialize external webcam)


while True:
    ret, frame = capt_entity.read()
    if not ret:
        break
    # save the frames to "result"
    
    result = model(frame)
    annotated_frame = results[0].plot()
    cv2.imshow("Using YoloV8", annotated_frame)
    
    if cv2.waitKey(1) & 0xFF == ord('e'):
        break

capt_entity.release()
cv2.destroyAllWindows()
```
